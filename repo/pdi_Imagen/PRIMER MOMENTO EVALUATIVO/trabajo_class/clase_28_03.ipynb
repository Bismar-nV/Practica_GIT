```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('daine.jpg',0)
t1=np.fft.fft2(img)
t2=np.fft.fftshift(t1)
fila, columna=img.shape
fila1, columna1=fila//2,columna//2

t2[fila-5:fila+5,columna1-5:columna1+5]=0
t3=np.fft.ifftshift(t2)
img_fin=np.fft.ifft2(t3)
img_fin=np.abs(img_fin)
espectro=20*np.log(np.abs(t2))


plt.figure(figsize=(15,10))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(img_fin, cmap='gray')
plt.title('fin Image')

plt.subplot(2, 2, 3)
plt.imshow(espectro, cmap='gray')
plt.title('espectro Image')


plt.show()
```
