```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('nami.jpg', cv2.IMREAD_GRAYSCALE)

histogram, _ = np.histogram(img.flatten(), 256, [0, 256])
cumulative_distribution = histogram.cumsum()
cumulative_distribution_normalized = cumulative_distribution * histogram.max() / cumulative_distribution.max()
img_equalized = cv2.equalizeHist(img)

plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(img_equalized, cmap='gray')
plt.title('Equalized Image')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('campo1.jpg')

histogram, _ = np.histogram(img.flatten(), 256, [0, 256])
funcion_dist = histogram.cumsum()
hst_normalizado = funcion_dist * histogram.max() / funcion_dist.max()
img_equalized = cv2.equalizeHist(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY))

plt.figure(figsize=(15,10))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(img_equalized, cmap='gray')
plt.title('Equalized Image')

plt.subplot(2, 2, 3)
plt.plot(funcion_dist, color='b')
plt.title('Cumulative Distribution')

plt.subplot(2, 2, 4)
plt.plot(hst_normalizado, color='r')
plt.title('Normalized Cumulative Distribution')

plt.show()
```
