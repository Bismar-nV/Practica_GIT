```python
import cv2 
import matplotlib.pyplot as plt
def histograma(imagen):
    histo=cv2.calcHist([imagen],[0],None,[256],[0,256])
    plt.figure(figsize=(15,8))
    plt.subplot(2,3,2)
    plt.plot(histo,color="blue")
    plt.title('HISTOGRAMA')
    plt.xlabel('intensidad_pixel')
    plt.ylabel('frecuencia')

    #ESTE ES UMBRALIZACION BINARIA
    _,umbralbi=cv2.threshold(img,127,255, cv2.THRESH_BINARY)
    _,umbralbi_inc_bi=cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
    umb_adap = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
    umb_adap_gauss = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

    plt.subplot(2,3,1)
    plt.imshow(img, cmap='gray')
    plt.title('imagen inicial')

    plt.subplot(2,3,3)
    plt.imshow(umbralbi, cmap='gray')
    plt.title('bianria')

    plt.subplot(2,3,4)
    plt.imshow(umbralbi_inc_bi, cmap='gray')
    plt.title('BInaRIO INVERSA')

    plt.subplot(2,3,5)
    plt.imshow(umb_adap, cmap='gray')
    plt.title('Umbralizacion adaptable')

    plt.subplot(2,3,6)
    plt.imshow(umb_adap_gauss, cmap='gray')
    plt.title('umbrali adaptable gaus')
    plt.show()

img=cv2.imread('nami.jpg',0)
histograma(img)
```
