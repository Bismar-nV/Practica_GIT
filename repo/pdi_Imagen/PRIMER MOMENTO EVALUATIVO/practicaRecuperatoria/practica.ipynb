```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('imagen2.jpg', cv2.IMREAD_GRAYSCALE)
alto, ancho = img.shape[:2]
mascara = np.zeros_like(img)
triangulo = np.array([[ancho//2, alto//2], [ancho//2 + 500, alto//2 - 500], [ancho//2 - 500, alto//2 - 500]])
cv2.fillPoly(mascara, [triangulo], 255)
img_fin = cv2.bitwise_and(img, img, mask=mascara)

plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.title('original')
plt.imshow(cv2.cvtColor(cv2.imread('imagen2.jpg'), cv2.COLOR_BGR2RGB))

plt.subplot(1, 3, 2)
plt.title('img en Triangulo')
plt.imshow(cv2.cvtColor(img_fin, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img= cv2.imread('imagen2.jpg', cv2.IMREAD_GRAYSCALE)
alto, ancho = img.shape[:2]
mascara = np.zeros_like(img)
num_circulos = 5
for i in range(num_circulos):
    centro = (np.random.randint(ancho), np.random.randint(alto))
    radio = np.random.randint(20, 130)
    color = np.random.randint(50, 200, size=3)
    cv2.circle(mascara, centro, radio, color.tolist(), -1)
img_fin = cv2.bitwise_and(img, img, mask=mascara)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Imagen con crculos de radn')
plt.imshow(cv2.cvtColor(img_fin, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


colores = [cv2.COLORMAP_AUTUMN, cv2.COLORMAP_BONE, cv2.COLORMAP_COOL,
           cv2.COLORMAP_HOT, cv2.COLORMAP_HSV, cv2.COLORMAP_JET,
           cv2.COLORMAP_OCEAN, cv2.COLORMAP_PARULA, cv2.COLORMAP_PINK,
           cv2.COLORMAP_RAINBOW, cv2.COLORMAP_SPRING, cv2.COLORMAP_SUMMER,
           cv2.COLORMAP_WINTER]
img = np.zeros((300, 400), dtype=np.uint8)* colores
alto, ancho = img.shape[:2]
mascara = np.zeros_like(img)
num_circulos = 5

for i in range(num_circulos):
    centro = (np.random.randint(ancho), np.random.randint(alto))
    radio = np.random.randint(10, 36)
    cv2.circle(mascara, centro, radio, colores[i], -1)

img_fin = cv2.bitwise_and(img, img, mask=mascara)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Imagen con crculos de radn')
plt.imshow(img_fin, cmap="gray")
plt.show()
```
