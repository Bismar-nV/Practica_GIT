redimencionar imagen

```python
import cv2
import matplotlib.pyplot as plt

img= cv2.imread('cancer.jpg')
escale=0.5

img_redim=cv2.resize(img,None,fx=escale,fy=escale)

plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.title('original')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title('img redimencionado')
plt.imshow(cv2.cvtColor(img_redim,cv2.COLOR_BGR2RGB))
plt.show()
```

tralation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img= cv2.imread('cancer.jpg')
x=120
y=60

martriz_tras=np.float32([[1,0,x],[0,1,y]])
img_tras=cv2.warpAffine(img,martriz_tras,(img.shape[1],img.shape[0]))


plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.title('original')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title('img tralation')
plt.imshow(cv2.cvtColor(img_tras,cv2.COLOR_BGR2RGB))
plt.show()
```

filtro mascara FIGURA

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img= cv2.imread('cancer.jpg', cv2.IMREAD_GRAYSCALE)
alto, ancho =img.shape[:2]
mascara=np.zeros_like(img)
circulo=(ancho//2,alto//2)
radio=100
cv2.circle(mascara,circulo,radio,(255),-1)
img_fin=cv2.bitwise_and(img,img,mask=mascara)

plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.title('original')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title('img filtro mascara')
plt.imshow(img_fin, cmap="gray")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img= cv2.imread('cancer.jpg', cv2.IMREAD_GRAYSCALE)
alto, ancho =img.shape[:2]
mascara=np.zeros_like(img)
circulo=(ancho//2,alto//2)
radio=100
cv2.circle(mascara,circulo,radio,(255),-1)
img_fin=cv2.bitwise_and(img,img,mask=mascara)
img_fin1=img*(mascara/255)+(img//2)*(1-mascara/255)

plt.figure(figsize=(10,5))
plt.subplot(1, 3, 1)
plt.title('original')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

plt.subplot(1, 3, 2)
plt.title('img filtro mascara')
plt.imshow(img_fin, cmap="gray")

plt.subplot(1, 3, 3)
plt.title('img filtro mascara degradado')
plt.imshow(img_fin1, cmap="gray")
plt.show()
```
