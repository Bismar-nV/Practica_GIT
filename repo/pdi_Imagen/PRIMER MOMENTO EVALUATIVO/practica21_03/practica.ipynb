```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('nami.jpg')
img_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

def contraste(imagen, a=1.0, b=0):
    img_nueva = np.clip(a * imagen + b, 0, 255).astype(np.uint8)
    return img_nueva

img_contrastada = contraste(img_gris, 1.5, 50)  

plt.figure(figsize=(20, 10))

plt.subplot(131)
plt.title('Imagen inicial')
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))

plt.subplot(132)
plt.title('Imagen en escala de grises')
plt.imshow(img_gris, cmap='gray')

plt.subplot(133)
plt.title('Imagen con contraste')
plt.imshow(img_contrastada, cmap='gray')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('nami.jpg')
img_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

def contraste(imagen, a=1.0, b=0):
    img_nueva = np.clip(a * imagen + b, 0, 255).astype(np.uint8)
    return img_nueva

# Aplicar contraste a la imagen en escala de grises
img_contrastada_gris = contraste(img_gris, 1.5, 50)

# Aplicar contraste a la imagen en color
img_contrastada_color = np.zeros_like(imagen)
for i in range(3):
    img_contrastada_color[:,:,i] = contraste(imagen[:,:,i], 1.5, 50)

plt.figure(figsize=(10, 8))

plt.subplot(221)
plt.title('Imagen original')
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))

plt.subplot(222)
plt.title('Imagen con contraste (color)')
plt.imshow(cv2.cvtColor(img_contrastada_color, cv2.COLOR_BGR2RGB))

plt.subplot(223)
plt.title('Imagen en escala de grises')
plt.imshow(img_gris, cmap='gray')

plt.subplot(224)
plt.title('Imagen con contraste (gris)')
plt.imshow(img_contrastada_gris, cmap='gray')


plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('nami.jpg')


def contraste(imagen, alpha=1.0, beta=0):
    img_nueva = cv2.convertScaleAbs(imagen, alpha=alpha, beta=beta)
    return img_nueva

blue, green, red = cv2.split(imagen)
contraste_blue = contraste(blue, alpha=5, beta=1.8)
contraste_red = contraste(red, alpha=2, beta=1.6)
contraste_green = contraste(green, alpha=1.6, beta=10)

fin_img = cv2.merge((contraste_blue, contraste_green, contraste_red))


plt.figure(figsize=(20, 10))

plt.subplot(231)
plt.title('Imagen original')
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(232)
plt.title('Imagen con contraste (final)')
plt.imshow(cv2.cvtColor(fin_img, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```
